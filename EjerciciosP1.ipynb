**Ejercicio 1**

Crea una consulta SQL que regrese todos los atributos de la tabla DataProducto, pero solo con los registros de los clientes que son de Brasil (Brazil)

```python
#USAMOS LA TABLA DATAPRODUCTO
query1 = """
SELECT *
FROM DataProducto as DP
WHERE DP.pais = 'Brazil';
"""

# Run the query
df_q1 = sqldf(query1)
df_q1
```

**Ejericicio 2**

Escribe una consulta que calcule la suma de los precios de cada registro, considerando únicamente los pagos realizados con visa o visa-electron, cuyo precio sea superior a 900,000. La suma debe agruparse por pais

```python
#USAMOS LA TABLA DATAPRODUCTO
query2 = """
SELECT DP.pais, SUM(DP.precio)
FROM DataProducto as DP
WHERE DP.forma_pago IN ('visa','visa-electron')
AND DP.precio > 900000
GROUP BY 1;
"""

# Run the query
df_q2 = sqldf(query2)
df_q2
```

**Ejercicio 3**

Genera una consulta que regrese el precio promedio del vehiculo por tipo de carro y por genero, excluyendo los siguientes paises

* Ukraine (Ucrania)
* Russia (Rusia)

y considerando solo los tipos de carro:

* Audi
* Mazda
* Nissan

```python
#USAMOS LA TABLA DATAPRODUCTO
query3 = """
SELECT DP.tipo_carro,DC.genero, AVG(DP.precio) 
FROM DataProducto as DP
LEFT JOIN DataCliente as DC
ON DP.id_cliente = DC.id_cliente
WHERE DP.pais NOT IN ('Ukraine','Russia')
AND DP.tipo_carro IN ('Audi','Mazda','Nissan')
GROUP BY 2,1;
"""

# Run the query
df_q3 = sqldf(query3)
df_q3
```
